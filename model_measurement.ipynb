```python
import random
actual_data=[random.randint(0,1) for i in range(0,20)]
pred_data=actual_data[0:8]+[1,0,0,1]+actual_data[12:]

print(actual_data)
print(pred_data)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
tn,fp,fn,tp=confusion_matrix(actual_data,pred_data).ravel()
print(tn,fp,fn,tp)
```

```python
def accuracy(tp,tn,fp,fn):
    return (tp+tn)/(tp+tn+fp+fn)
```

```python
a=accuracy(tp,tn,fp,fn)
print(a)
a=accuracy_score(actual_data,pred_data)
print(a)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

cm=confusion_matrix(actual_data,pred_data)
graph=ConfusionMatrixDisplay(confusion_matrix=cm)
graph.plot()
```

```python
import pandas as pd
array=[[tp,fn],[fp,tn]]
df=pd.DataFrame(array)
import seaborn as sns
sns.heatmap(df,annot=True)
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
print(roc_auc_score(actual_data,pred_data))
```

```python
import matplotlib.pyplot as plt

fpr,tpr,_=roc_curve(actual_data,pred_data)

plt.title("roc curve")
plt.xlabel("fpr")
plt.ylabel("tpr")

plt.plot(fpr,tpr)
plt.plot([0,1],[0,1],linestyle="--")

plt.xlim([0,1.05])
plt.ylim([0,1.05])
```
